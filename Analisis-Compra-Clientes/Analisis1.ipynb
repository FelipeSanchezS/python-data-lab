#Limpieza de datos

```python
#muestrame cuantos null o nan hay
df.isnull().sum()
#Podemos ver que no hay ningun dato nulo
```

```python
#cambiar los datos nulos por 0
df.fillna(0)
```

```python
#Visualizar el total de filas duplicadas
print(df.duplicated().sum())
```

```python
#Eliminar4 filas duplicadas
df.drop_duplicates(inplace=True)
```

#Analisis de datos

```python
#¿Cuál es la distribución de edades de los clientes?

#Cual es el promedio de edades de los clientes
print("La media de edades es: ", df['Age'].mean())
#visualizar la distribución de edades de los clientes
sns.histplot(df['Age'], color='red')
plt.title('Distribución de edades de los clientes')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.show()
```

```python
#¿Qué género realiza más compras segun el dataset?

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title('Cantidad de Compras por Género')
plt.xlabel('Género')
plt.ylabel('Cantidad de Compras')
plt.show()
```

```python
#Grafica de pastel con las compras de hombres vs mujeres
labels = df['Gender'].value_counts().index
plt.pie(df['Gender'].value_counts(), labels=labels, autopct='%1.1f%%')
plt.title('Porcentaje de compras por género')
plt.show()
```

```python
#¿Cómo se distribuyen los métodos de pago utilizados?

#Distribición de metodos de pagos usados
plt.figure(figsize=(9, 5))
sns.countplot(data=df, x='Payment Method', palette='pastel')
plt.title('Cantidad de Compras por Método de Pago')
plt.xlabel('Método de Pago')
plt.ylabel('Cantidad de Compras')
plt.show()
```

```python
#¿Existe una relación entre la edad y el monto de la compra?

#Relacion entre edad y el monto de la compra
sns.scatterplot(data=df, x='Age', y='Purchase Amount (USD)', hue='Gender', palette='pastel')
plt.title('Relación entre Edad y Monto de Compra')
plt.xlabel('Edad')
plt.ylabel('Monto de Compra (USD)')
plt.show()
```

```python
#¿Cuál es la temporada con más compras?

#Season donde mas compras se realizan
sns.countplot(data=df, x='Season', palette='pastel')
plt.title('Cantidad de Compras por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Cantidad de Compras')
plt.show()
```

```python
#¿Cules con los articulos que mas compra la gente?

#Frecuencia del top 5 de los articulos mas comprados
top_5_items = df['Item Purchased'].value_counts().head(5)
sns.barplot(x=top_5_items.index, y=top_5_items.values, palette='pastel')
plt.title('Top 5 de los artículos más comprados')
plt.xlabel('Artículo')
plt.ylabel('Cantidad de Compras')
```

```python
#Top 5 de los articulos con mejor calificación de revisión
top_5_items = df.sort_values('Review Rating', ascending=False).head(5)
sns.barplot(x=top_5_items['Item Purchased'], y=top_5_items['Review Rating'], palette='pastel')
```

```python
#con que frecuencia las personas realizan compras
plt.figure(figsize=(11, 5))
sns.countplot(data=df, x='Frequency of Purchases', palette='pastel')
plt.title('Frecuencia de Compras')
plt.xlabel('Frecuencia')
plt.ylabel('Cantidad de Compras')
```

```python
#tipo de envio con mayor monto promedio de compra
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Shipping Type', y='Purchase Amount (USD)', palette='pastel')
plt.title('Tipo de Envío con Mayor Monto Promedio de Compra')
plt.xlabel('Tipo de Envío')
```

```python
#¿Top 5 articulos con peor recomendación?

top_5_items_bad = df.sort_values('Review Rating', ascending=True).head(5)
sns.barplot(x=top_5_items_bad['Item Purchased'], y=top_5_items_bad['Review Rating'], palette='pastel')
plt.title('Top 5 de los artículos con peor recomendación')
plt.xlabel('Artículo')
plt.ylabel('Calificación de revisión')
```
