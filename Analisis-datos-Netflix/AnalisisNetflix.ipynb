##Visualizaciones de datos

```python
#cantidad de tv show y de movie en el dataset
df['type'].value_counts()
```

```python
#¿Cuál es la distribución en % de tv show y movie en el dataset?
plt.pie(df['type'].value_counts(), labels=df['type'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribución de TV Shows y Movies')
plt.show()
```

```python
#¿Cuáles son los países con más títulos en el dataset?
info = df['country'].value_counts().head(5)
print(info)
df['country'].value_counts().head(5).plot(kind='bar')
plt.title('Paises con mas titulos')
plt.xlabel('Paises')
plt.ylabel('Cantidad de titulos')
plt.show()
```

```python
#Como ha evlucionado la csntidad de titulos a lo largo de los años
#convertimos la variable date_added a datetime
df['date_added'] = pd.to_datetime(df['date_added'])
avance=df['year_added'] = df['date_added'].value_counts()
print(avance)
```

```python
#Cuales son los 3 años con mas lanzamientos de peliculas o series
df['release_year'].value_counts().head(3).plot(kind='bar')
plt.title('Años con mas lanzamientos')
plt.xlabel('Años')
plt.ylabel('Cantidad de titulos')
plt.show()
```

```python
#distribución de ranting entre peliculas y series
df['rating'].value_counts().plot(kind='bar')
plt.title('Ranting')
plt.xlabel('Ranting')
plt.ylabel('Cantidad de titulos')
plt.show()
```

```python
#Distribución en la duración de peliculas y filtramos poy type == Movie
df_movie = df[df['type'] == 'Movie']
df_movie['duration'].value_counts().plot(kind='bar')
plt.title('Duracion')
plt.xlabel('Duracion')
plt.ylabel('Cantidad de titulos')
plt.show()
```

```python
#Cuales son los 10 titulos mas largos
largas=df.sort_values(by='duration', ascending=False).head(10)
cortos=df.sort_values(by='duration', ascending=True).head(10)
print(largas)
print(cortos)
```

```python
#directores con mas firmaciones
df['director'].value_counts().head(10)
```

```python
#Los titulos añadidos mas recientemente
df.sort_values(by='date_added', ascending=False).head(10)
```

```python
#cuales son los generos mas comunes entre series y peliculas
df['listed_in'].value_counts().head(10)
```

```python
#Agrupación de ranting por serie y peliculas
df.groupby(['type', 'rating']).size()
```

```python
#graficamos lo anterior
df.groupby(['type', 'rating']).size().plot(kind='bar')
plt.title('Ranting')
plt.xlabel('Ranting')
plt.ylabel('Cantidad de titulos')
plt.show()
```

```python
#como varia la duración promedio de las peliculas sun su ranting
df_movie = df[df['type'] == 'Movie']
# Extract numeric duration and convert to integers
df_movie['duration'] = df_movie['duration'].str.extract('(\d+)').astype(int)
# Calculate the average duration and plot it as a bar chart using plt.bar
avg_duration = df_movie['duration'].mean()
plt.bar(['Average Duration'], [avg_duration])  # Plot a single bar for average duration
plt.title('Duracion')
plt.xlabel('Duracion')
plt.ylabel('Cantidad de titulos')
plt.show()
```

```python
#cuantos titulos fueron lanzados en cada año
df['release_year'].value_counts().plot(kind='bar', figsize=(15, 5))
plt.title('Años con mas lanzamientos')
plt.xlabel('Años')
plt.ylabel('Cantidad de titulos')
plt.show()
```
