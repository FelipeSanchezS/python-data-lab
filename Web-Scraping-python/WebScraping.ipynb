```python
#ingresamos la url a la que queremos hacer web scraping
website ="https://www.vulnhub.com/"
resultado = requests.get(website)
```

```python
content = resultado.text
print(content)
```

```python
#empezamos a seccionar las partes que nos interesan del sitio
import re

# Suponiendo que 'content' es el HTML o el texto del cual estás extrayendo la información
patron = r"/entry/([^/]+)"
maquinas_repetidas = re.findall(patron, str(content))
print(maquinas_repetidas)
```

```python
#lo ponemos de mejor forma visual
#liminamos elementos repetidos
sin_duplicados = list(dict.fromkeys(maquinas_repetidas))
print(sin_duplicados)
```

```python
maquinas_final = []
for i in sin_duplicados:
    nombre_maquinas = i.replace("/entry/", "")  # Replace "/entry/" with an empty string
    maquinas_final.append(nombre_maquinas)
print(nombre_maquinas)
```

```python
#indicacion si hay cambios en la pagina  o no
maquina_noob = "noob-1,746"
existe_noob = False
for maquina in maquinas_final:
    if maquina == maquina_noob:
        existe_noob = True
        break
```

```python
#aplicamos estilos

color_verde = Fore.GREEN
color_amarillo = Fore.YELLOW
```

```python
#validamos la condicion de existencia
if existe_noob:
    print(color_verde + f"La maquina {maquina_noob} existe en la lista, no hay cambios.")
else:
    print(color_amarillo + f"La maquina {maquina_noob} no existe en la lista, es decir, hay cambios.")
```
