##Tratamiento y limpieza de datos

```python
df.isnull().sum()
```

```python
#validamos en tabla true false valorews null
df.isnull()
```

```python
#Eliminamos las filas donde existan valores null
df.dropna(subset=['ratings'], inplace=True)
df.dropna(subset=['no_of_ratings'], inplace=True)
df.dropna(subset=['discount_price'], inplace=True)
df.dropna(subset=['actual_price'], inplace=True)
```

```python
#validamos nuevamente datos nulos
df.isnull().sum()
```

```python
df.isna()
```

##Analisis y visualización de datos

```python
#convertidos la cxolumna actual_price a int
# Remove currency symbols and commas before converting to integers
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(int)
```

```python
df
```

```python
#hallamos la media del valor actual de los aires
df['actual_price'].mean()
```

```python
df['discount_price'] = df['discount_price'].str.replace('₹', '').str.replace(',', '')
df['discount_price'] = df['discount_price'].astype(int)
```

```python
#hallamos el promedio de los descuentos
df['discount_price'].mean()
```

```python
#relacion entre actual_ price y rantings
sns.scatterplot(x='actual_price', y='ratings', data=df)
plt.show()
```

```python
#validamos la cantidad de calificaciones de aires acondicionales
df['ratings'].value_counts()
```

```python
plt.pie(df['ratings'].value_counts(), labels=df['ratings'].value_counts().index, autopct='%1.1f%%')
plt.title('Porcentaje de calificaciones')
plt.show()
```

```python
#conteo de rantings segun so ponderado
df['ratings'].value_counts().plot(kind='bar')
plt.title('Conteo de Ranting')
plt.xlabel('Ranting')
plt.ylabel('Cantidad de calificaciones')
plt.show()
```

```python
#los 5 productos con peor rating
df.sort_values(by='ratings', ascending=False).head(5)
```

```python
#graficamos lo antetior
sns.barplot(x='ratings', y='name', data=df.sort_values(by='ratings', ascending=False).head(5))
plt.show()
```

```python
#los 5 productos con mejor rantings
df.sort_values(by='ratings', ascending=True).head(5)
```

```python
#El nombre de los 5 productos con mayor descuento
df.sort_values(by='discount_price', ascending=False).head(5)
```

```python
#El nombre de los 5 productos con menor descuento
df.sort_values(by='discount_price', ascending=True).head(5)
```

```python
#graficamos la antetior consulta
sns.barplot(x='discount_price', y='name', data=df.sort_values(by='discount_price', ascending=True).head(5))
plt.show()
```

```python
#relacion entre precio y discount
sns.scatterplot(x='actual_price', y='discount_price', data=df)
plt.show()
```
