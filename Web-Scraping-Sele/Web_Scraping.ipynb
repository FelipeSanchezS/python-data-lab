```python
import undetected_chromedriver as uc

options = uc.ChromeOptions()
options.add_argument('--headless')  # Ejecuta sin interfaz gráfica
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

browser = uc.Chrome(options=options)
url = "https://carro.mercadolibre.com.co/MCO-1582354049-mercedes-benz-a200-13-progressive-_JM#polycard_client=search-nordic&position=1&search_layout=grid&type=item&tracking_id=311d9275-4052-4cfa-b2c4-7e50e6cb3bef"
browser.get(url)

print("Título de la página:", browser.title)
```

```python
#le decimos que acepte las cookies
browser.find_element("xpath", '//*[@class="cookie-consent-banner-opt-out__action cookie-consent-banner-opt-out__action--primary cookie-consent-banner-opt-out__action--key-accept"]').click()
```

```python
#Sacamos el HTML de la pagina
html = browser.page_source
```

```python
html
```

```python
soup = bs(html, 'lxml')
soup
```

```python
#Para extraer la información debemos ver como esta estructurado el html y extraer el ID o class de como se armo

#Extraemos el titulo de la pagina
titulo = soup.find('h1', class_='ui-pdp-title').text
print("El titulo del carro es: "+titulo)
#Extraemos la Precio de la pagina
precio = soup.find('span', class_='andes-money-amount__fraction').text
print("El precio es: "+precio)
#Extraemos el Kilometraje del vehiculo
km = soup.find('h2', class_="ui-pdp-color--BLACK ui-pdp-size--MEDIUM ui-pdp-family--REGULAR ui-pdp-price-comparison__title-legend").text
print('El kilometraje es: '+km)
```

```python
#Con este element visualizamos el HTML
#print(soup.prettify())  # Para ver el contenido real
```

```python
#Extraemos caracterisicas del carro
caracteristicas = {}

parent = soup.find('tbody', {'class': "andes-table__body"})

#Aca recorremos cada uno de los elementos de la tabla y se crea un arreglo para que este guarde los elementos
if parent:
    for row in parent.find_all('tr'):
        columnas = row.find_all(['th', 'td'])
        if len(columnas) == 2:
            clave = columnas[0].get_text(strip=True)
            valor = columnas[1].get_text(strip=True)
            caracteristicas[clave] = valor

for clave, valor in caracteristicas.items():
    print(f"{clave}: {valor}")
```

```python
#Extraemos caracterisicas 2 del carro
import re

parent = soup.find('div', {'class': "ui-pdp-collapsable__container"}).text.strip()

# Insertar salto de línea antes de cada letra mayúscula, excepto la primera
texto_formateado = re.sub(r'(?<!^)(?=[A-Z])', '\n', parent)

print(texto_formateado)
```

```python
# Encuentra todas las ocurrencias del div con esa clase
ubicaciones = soup.find_all('div', {'class': "ui-pdp-media__body"})

# Verifica que haya al menos dos ocurrencias
if len(ubicaciones) >= 7:
    ubi = ubicaciones[6].text.strip()
    print("La ubicación del vehiculo es: "+ubi)
```
