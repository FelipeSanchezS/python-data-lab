```python
#Importamos librerias numpy y tensorflow
import numpy as np
import tensorflow as tf
```

```python
#Creamos arreglo con los datos de entrada, uno para celsius y otro para fahrenheit
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
#estructuramos la red neuronal mediante keras
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
#Compilar modelo para ser entrenado, usamos el optimizador y la función de error
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
print("Comenzando entrenamiento...")
historial = modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Modelo entrenado!")
```

```python
#Validamos la perdida de los datos
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
#Una vez entrenado haremos una predicción, en este caso con 100 celsius
resultado = modelo.predict(np.array([100.0])) # Changed this line to pass a NumPy array
print("El resultado es " + str(resultado) + " fahrenheit!")
```
